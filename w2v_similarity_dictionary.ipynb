```python
from gensim.models import Word2Vec

model = Word2Vec.load('./skipgram_model_10_20.model')
```

```python
hypernyms = ['카페', '술집', '노포', '맛집', '소품샵', '복합', '서점', '전시', '편집샵',
'유적', '캠핑', '등산', '관광지', '명소', '촬영지', '체험', '원데이', '혼자', '혼커', '혼밥', '혼술',
'우정', '럽스타', '부부', '가족', '육아', '한옥', '사진', '출사', '스냅', '뷰가', '뷰맛집', '운치', '노을', '산책',
'테라스', '루프탑', '디저트', '애견', '비건', '빈티지한', '모던한', '드라이브', '뚜벅이']
```

```python
# hyponyms
hyponyms = []
for hypernym in hypernyms:
    hyponyms.append(model.wv.most_similar(hypernym, topn = 300))
```

```python
import pandas as pd

df = pd.DataFrame({'hypernyms': hypernyms, 'hyponyms': hyponyms})
df.to_csv('./hypernyms_hyponyms.csv', index=False, encoding='utf-8')
df.head()
```
