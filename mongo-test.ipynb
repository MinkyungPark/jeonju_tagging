```python
from collections import defaultdict
import tqdm

for loc_pk, loc_name, num_docu, keywords_score, raw in df.values.tolist():
    push_dic = {
        '_id': loc_pk,
        'sns_name': loc_name,
        'num_of_documents': num_docu,
        'raw_keywords_score': keywords_score,
        'tags_score': {},
        'tags_hyponym': {},
    }
    tag_score = defaultdict(float)
    tag_hyponym = defaultdict(dict)
    for tag, hyponym in tags_hyponyms_dic.items():

        for (word, score) in keywords_score:
            for hypo_word in hyponym:

                if hypo_word in word:
                    tag_score[tag] += score
                    tag_hyponym[tag][hypo_word] = score



    tag_score = dict(sorted(tag_score.items(), key=lambda x: -x[1]))
    if tag_score:
        push_dic['tags_score'] = tag_score
        push_dic['tags_hyponym'] = tag_hyponym
        push_dic['top_tags'] = list(tag_score.keys())[:3]
        collection.insert_one(push_dic)
        print(f'{loc_pk} db push done!')
        
```
