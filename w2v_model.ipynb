```python
import pickle
from gensim.models.word2vec import Word2Vec
```

```python
with open('./all_content_hashtag.pkl', 'rb') as f:
    all_content_hashtag = pickle.load(f)
```

```python
corpus = []
for pk, dic in all_content_hashtag.items():
    corpus.append(dic['join'].split())
len(corpus)
```

```python
corpus = [c for c in corpus if c]
len(corpus)
```

```python
# 한글자 제외
corpus = [[y for y in x if not len(y) == 1] for x in corpus]
```

```python
model = Word2Vec(
    corpus,
    sg = 1, # skim gram
    window = 5, # 중심 단어로부터 좌우 5 단어까지 학습
    min_count = 10, # 최소 등장 횟수
)

model.init_sims(replace=True)
```

```python
model.wv.similarity('커피', '갤러리')
```

```python
model.wv.most_similar('카페', topn=100)
```

```python
model3 = Word2Vec(
    corpus,
    sg = 1, # skim gram
    window = 5, # 중심 단어로부터 좌우 5 단어까지 학습
    min_count = 50, # 최소 등장 횟수
)

model3.init_sims(replace=True)
```
